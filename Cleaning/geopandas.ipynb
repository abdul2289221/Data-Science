### Reading the JSON file with the JSON package

```python
# Importing the needed JSON library
import json
```

```python
# Opening the file (for some reason I don't know, it won't work with the default encoding='utf-8')
with open('current_city_list.json', encoding='latin-1') as file:
    data = json.load(file)

# data
```

- The data is a list of dict

```python
# Let's check the first element
data[-1]
```

### Restructuring the data

- What we want is: 'id', 'lon', 'lat', 'country', 'name', 'population'

```python
# Realizing on the 1rst element of the data
# city_dict = {}
# city_dict['id'] = data[0]['id']
# city_dict['country'] = data[0]['country']
# city_dict['name'] = data[0]['name']
# city_dict['population'] = data[0]['stat']['population']
# city_dict['lon'] = data[0]['coord']['lon']
# city_dict['lat'] = data[0]['coord']['lat']
```

```python
# Generalizing to the whole data
city_dict = {'id':[], 'country':[], 'name':[], 'population':[], 'lon':[], 'lat':[]}
 
for i in range(len(data)):
    city_dict['id'].append(data[i]['id'])
    city_dict['country'].append(data[i]['country'])
    city_dict['name'].append(data[i]['name'])
    city_dict['population'].append(data[i]['stat']['population'])
    city_dict['lon'].append(data[i]['coord']['lon'])
    city_dict['lat'].append(data[i]['coord']['lat'])

# city_dict
```

```python
# DataFrame
df1 = pd.DataFrame(city_dict)
df1
```

Now we have our pretty DataFrame

```python
# Let's chech how many countries we got in the dataset
df1['country'].nunique()
df1['country'].count()
```

```python
# Let's group by country to see 
pop_per_city = df1.groupby(by=['country'])[['population']].sum().sort_values(by='population', ascending=False)
pop_per_city.reset_index().head(30)
```

```python
# Top 10 country with the most population
top_10_country = pop_per_city.head(10)
top_10_country
```

```python
# 
top_10_country.index.to_list()
```

```python
mask = df1['country'].isin(top_10_country.index.to_list())
df2 = df1[mask]
# mask
df2
```

### Plotting the Top 10 Countries with the most Populations Cities

```python
# ! pip install geopandas
# ! pip install leafmap
```

```python
# Importing Geopandas and Leafmap 
import geopandas as gpd
import leafmap
```

```python
# 
g_df = gpd.GeoDataFrame(data=df2, geometry=gpd.points_from_xy(df2['lat'], df2['lon']), crs='EPSG:4326')
g_df
```

```python
# Using Leafmap to plot the map
m = leafmap.Map()
m.add_basemap(basemap='Google Maps')
m.add_gdf(g_df.iloc[:50], layer_name='Top 10 Countries')
m.add_title('Cities of Top 10 Countries with the most Population', align='center', font_size='20px')
```
