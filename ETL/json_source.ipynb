```python
# Dropping unnecessary columns
df.drop(columns=['page 1 (main category)', 'page 2 (clothing model)', 'model photography', 'page'], inplace=True)
df.head()
```

#### Top 20 countries with the most orders

```python
# # data_df['Order'].mean()
# #data_df.groupby('Country')[['Order', 'Price']].mean().sort_values('Order', ascending=False)
# grp_country = data_df.groupby('Country')[['Order']].sum()
# #list(grp_country)
# #grp_country
# top_20_countries = grp_country.sort_values('Order', ascending = False)
# top_20_countries = top_20_countries.reset_index().head(20)
# top_20_countries.head()
```

```python
# Total number of orders by country
country_orders = df.groupby(['country'])[['order']].sum().reset_index()
country_orders.head()
```

```python
# 
top_20_countries = country_orders.sort_values(by='order', ascending=False).head(20)
top_20_countries
```

### Loading to CSV file

```python
# 
top_20_countries.to_csv('top_20_countries_with_most_order.csv', index = False)
```
